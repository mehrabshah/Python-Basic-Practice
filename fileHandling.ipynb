```python
word_stats={}

with open('poem.txt', 'r') as file:

    for line in file :
       words=line.split(' ')
       for word in words:
          if word in word_stats:
             word_stats[word] += 1
          else:
             word_stats[word] = 1

for keys,values in word_stats.items():
   print(f"{keys}=>{values}")



```
